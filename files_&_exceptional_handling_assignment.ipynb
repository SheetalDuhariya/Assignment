```python
# 4.  Write a Python script that reads from one file and writes its content to another file

def copy_file_line_by_line(source_file, destination_file):
    try:
        with open(source_file, 'r') as source:
            with open(destination_file, 'w') as destination:
                # Read and write line by line
                for line in source:
                    destination.write(line)
        print(f"File successfully copied from {source_file} to {destination_file}")

    except FileNotFoundError:
        print(f"Error: Source file {source_file} not found.")
    except PermissionError:
        print("Permission denied to read or write files.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")
```

```python
# 6. Write a Python program that logs an error message to a log file when a division by zero exception occurs

import logging
import traceback

def advanced_logging_division():
    """
    Advanced error logging with full traceback
    """
    logging.basicConfig(
        filename='advanced_division_errors.log',
        level=logging.ERROR,
        format='%(asctime)s - %(levelname)s\n%(message)s',
        datefmt='%Y-%m-%d %H:%M:%S'
    )

    def safe_division_with_traceback(a, b):
        try:
            return a / b
        except ZeroDivisionError:
            # Log full traceback for comprehensive error details
            logging.error(
                f"Division Error Details:\n"
                f"Numerator: {a}\n"
                f"Denominator: {b}\n"
                f"Traceback:\n{traceback.format_exc()}"
            )
            return None

    # Test scenarios
    test_cases = [(10, 2), (5, 0), (15, 3)]
    for num, denom in test_cases:
        result = safe_division_with_traceback(num, denom)

```

```python
# 18. How would you implement a basic logging setup that logs to a file with rotation after 1MB

import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)  # Set the logging level to DEBUG

# Create a rotating file handler that logs to 'app.log'
handler = RotatingFileHandler('app.log', maxBytes=1 * 1024 * 1024, backupCount=5)
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example log messages
for i in range(10000):
    logger.debug(f"This is log message number {i}")  # Log messages to fill the file

print("Logging setup complete. Check 'app.log' for log messages.")
```

```python
# 19. Write a program that handles both IndexError and KeyError using a try-except block

def access_data(data_list, data_dict, list_index, dict_key):
    try:
        # Attempt to access an element in the list
        list_value = data_list[list_index]
        print(f"Value from list at index {list_index}: {list_value}")

        # Attempt to access a value in the dictionary
        dict_value = data_dict[dict_key]
        print(f"Value from dictionary for key '{dict_key}': {dict_value}")

    except IndexError:
        print(f"Error: Index {list_index} is out of range for the list.")

    except KeyError:
        print(f"Error: Key '{dict_key}' not found in the dictionary.")

def main():
    # Sample data
    sample_list = [10, 20, 30, 40, 50]
    sample_dict = {'a': 1, 'b': 2, 'c': 3}

    # Test cases
    test_cases = [
        (2, 'b'),   # Valid index and key
        (5, 'a'),   # Invalid index and valid key
        (1, 'd'),   # Valid index and invalid key
        (10, 'e')   # Invalid index and invalid key
    ]

    for index, key in test_cases:
        print(f"\nAccessing index {index} and key '{key}':")
        access_data(sample_list, sample_dict, index, key)

if __name__ == "__main__":
    main()
```

```python
# 21 Write a Python program that reads a file and prints the number of occurrences of a specific word

def count_word_occurrences(filename, target_word):
    """Count occurrences of a specific word in a file."""
    try:
        with open(filename, 'r') as file:
            content = file.read()  # Read the entire content of the file

            # Normalize the content to lowercase for case-insensitive counting
            content = content.lower()
            target_word = target_word.lower()  # Normalize target word to lowercase

            # Count occurrences of the target word
            word_count = content.split().count(target_word)

            print(f"The word '{target_word}' occurs {word_count} times in the file '{filename}'.")

    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

def main():
    # Prompt user for filename and word to search
    filename = input("Enter the filename: ")
    target_word = input("Enter the word to count: ")

    # Count occurrences of the specified word
    count_word_occurrences(filename, target_word)

if __name__ == "__main__":
    main()
```

```python
# 23 Write a Python program that writes to a log file when an error occurs during file handling.

import logging

# Configure the logging
logging.basicConfig(
    filename='file_handling_errors.log',  # Log file name
    level=logging.ERROR,                   # Set the logging level to ERROR
    format='%(asctime)s - %(levelname)s - %(message)s'  # Log message format
)

def read_file_contents(filename):
    """Read and print the contents of a file."""
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(content)  # Print the content if read successfully
    except FileNotFoundError as e:
        logging.error(f"File not found: {e}")  # Log error if file is not found
        print(f"Error: The file '{filename}' does not exist.")
    except IOError as e:
        logging.error(f"I/O error occurred: {e}")  # Log I/O errors
        print(f"An I/O error occurred while reading the file: {e}")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")  # Log any other exceptions
        print(f"An unexpected error occurred: {e}")

def main():
    # Prompt user for the filename to read
    filename = input("Enter the filename to read: ")

    # Attempt to read the contents of the specified file
    read_file_contents(filename)

if __name__ == "__main__":
    main()
```
