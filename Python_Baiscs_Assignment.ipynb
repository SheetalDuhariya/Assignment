```python
# Que 3. Compare and contrast mutable and immutable objects in Python with examples


# Mutable Objects: These are objects whose internal state can be changed after they are created. This means you can modify, add, or remove elements without creating a new object.

# LISTS :

my_list = [1, 2, 3]
my_list[0] = 4  # List now becomes [4, 2, 3]

# Dictionaries :

my_dict = {'name': 'Alice', 'age': 25}
my_dict['age'] = 26  # Dictionary now becomes {'name': 'Alice', 'age': 26}

# SETS :

my_set = {1, 2, 3}
my_set.add(4)  # Set now becomes {1, 2, 3, 4}


# Immutable Objects: These objects cannot be altered once they are created. Any modification results in the creation of a new object.

#  Integers :

num = 10
num += 1  # Creates a new integer object with value 11

# Strings :

my_string = "Hello"
my_string += " World"  # Creates a new string object "Hello World"

# Tuples :

my_tuple = (1, 2, 3)
# Attempting to change an element will raise an error


```

```python
# Que 4 . Discuss the different types of operators in Python and provide examples of how they are used

# Arithmetic Operators - Addition , Subtraction , Subtraction , Division , Floor Division , Modulo , Exponentiation

a = 7
b = 2
print('Sum:', a + b)          # Output: Sum: 9
print('Subtraction:', a - b)  # Output: Subtraction: 5
print('Multiplication:', a * b) # Output: Multiplication: 14
print('Division:', a / b)      # Output: Division: 3.5
print('Floor Division:', a // b) # Output: Floor Division: 3
print('Modulo:', a % b)        # Output: Modulo: 1
print('Power:', a ** b)        # Output: Power: 49

# Assignment Operators - "=" "+=" "-=" "*=" "/=" "%=" "**="

a = 10
b = 5
a += b   # Now a is 15
print(a) # Output: 15

# Comparison Operators - "==" "!=" "<" ">" "<=" ">="

a = 10
b = 20
print(a == b) # Output: False
print(a != b) # Output: True

# Logical Operators - and, or, not

x = True
y = False
print(x and y) # Output: False
print(x or y)  # Output: True
print(not x)   # Output: False

# Bitwise Operators -  {&	Bitwise AND	(x & y)}  {`	`	Bitwise OR } {^	Bitwise XOR	(x ^ y)} {~	Bitwise NOT}	{(~x) <<	Left shift	(x << n)} {>>	Right shift	(x >> n)}

a = 60 # Binary: 0011 1100
b = 13 # Binary: 0000 1101

print(a & b) # Output: 12 (Binary: 0000 1100)
print(a | b) # Output: 61 (Binary: 0011 1101)

# Identity Operators - Identity operators are used to check if two variables refer to the same object in memory.
# is	Identity check	(a is b) ,, is not	Non-identity check	(a is not b)

a = [1,2,3]
b = a
c = list(a)

print(a is b)    # Output: True (both refer to the same list)
print(a is c)    # Output: False (c is a different object)

# Membership Operators -  Membership operators are used to test if a value exists within an iterable.
# in	Checks membership	(x in list) ,, not in	Checks non-membership.	(x not in list)

list1 = [1, 2, 3, 4]
print(3 in list1)    # Output: True
print(5 not in list1) # Output: True


```

```python
# Que 5 . Explain the concept of type casting in Python with examples

# Type casting in Python refers to the conversion of one data type into another. This process is essential for manipulating data effectively, especially when dealing with user inputs or performing operations that require specific data types. Python supports two main types of type casting: implicit and explicit.

# Implicit Type Casting - Implicit type casting occurs automatically when Python converts a smaller data type to a larger data type without any user intervention. This typically happens in arithmetic operations where the operands are of different types.

a = 10      # Integer
b = 5.5     # Float
c = a + b   # Implicitly converts 'a' to float
print(c)    # Output: 15.5

# Explicit Type Casting -  Explicit type casting, on the other hand, requires the programmer to specify the conversion using built-in functions. Common functions for explicit casting include int(), float(), and str(). Here are some examples:

# Converting String to Integer
num_str = "123"
num_int = int(num_str)  # Converts string to integer
print(num_int)           # Output: 123

# Converting Integer to Float
num_int = 10
num_float = float(num_int)  # Converts integer to float
print(num_float)             # Output: 10.0

# Converting Float to String
num_float = 3.14
num_str = str(num_float)  # Converts float to string
print(num_str)            # Output: '3.14'

```

```python
# Que 6 . How do conditional statements work in Python? Illustrate with examples

# 1. The if Statement - The if statement evaluates a condition, and if that condition is true, it executes a block of code.
num = 5

if num > 0:
    print("The number is positive.")

# 2. The else Statement -  The else statement provides an alternative block of code that executes if the if condition is false.
num = -3

if num > 0:
    print("The number is positive.")
else:
    print("The number is negative.")

# 3. The elif Statement -  The elif (short for "else if") statement allows you to check multiple conditions sequentially. You can have multiple elif statements following an if.
num = 0

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# 4. Nested Conditional Statements

num = 10

if num >= 0:
    if num == 0:
        print("The number is zero.")
    else:
        print("The number is positive.")
else:
    print("The number is negative.")

```

```python
# Que 7. Describe the different types of loops in Python and their use cases with examples.

# For Loop - The for loop in Python is used to iterate over a sequence (like a list, tuple, dictionary, set, or string). It allows you to execute a block of code repeatedly for each item in the sequence.

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

for i in range(5):
    print(i)  # prints numbers from 0 to 4

#  Nested For Loops: You can nest one for loop inside another to handle multi-dimensional data.

colors = ["red", "green", "blue"]
shapes = ["circle", "square"]

for color in colors:
    for shape in shapes:
        print(color, shape)

# Break and Continue Statements

# Break: Exits the loop prematurely.
for fruit in fruits:
    if fruit == "banana":
        break
    print(fruit)  # prints apple only

#  Continue: Skips the current iteration and continues with the next.

for fruit in fruits:
    if fruit == "banana":
        continue
    print(fruit)  # prints apple and cherry only

# While Loop - The while loop continues to execute as long as a specified condition is true. This type of loop is useful when the number of iterations is not known beforehand.
count = 5
while count > 0:
    print(count)
    count -= 1

count = 5
while count > 0:
    if count == 3:
        break   # Exit when count is 3
    print(count)
    count -= 1

# Output will be:
# 5
# 4


```
