```python
# Que 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days

import plotly.express as px
import pandas as pd
import numpy as np
import seaborn as sns

np.random.seed(20)
data = {
'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
'Day': np.random.choice(range(1, 31), 100),
'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table to aggregate sales by month and day
sales_pivot = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')

# Generate the heatmap using the pivot table
sns.heatmap(sales_pivot, annot=True, cmap='coolwarm', fmt=".1f")
#annot=True displays the values in each cell
#fmt=".1f" formats the displayed values to one decimal place
plt.title('Sales Heatmap')
plt.show()
```

```python
# Que 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.palettes import Viridis256

# Set random seed for reproducibility
np.random.seed(42)

# Generate random data
num_points = 100
x = np.random.random(num_points)
y = np.random.random(num_points)
sizes = np.random.randint(10, 50, num_points)
colors = np.random.choice(Viridis256, num_points)

# Create output file
output_file("random_scatter.html")

# Create figure with customized properties
p = figure(title='Random Scatter Plot',
           x_axis_label='X Values',
           y_axis_label='Y Values',
           width=600,
           height=400)

# Add scatter plot with variable sizes and colors
p.scatter(x, y,
          size=sizes,
          color=colors,
          alpha=0.7)

# Show the plot
show(p)



```

```python
# Que 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]

import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.palettes import Spectral10
from bokeh.models import ColumnDataSource

# Create fruit dataset
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
counts = [45, 30, 22, 15, 10]

# Create ColumnDataSource
source = ColumnDataSource(data=dict(
    fruits=fruits,
    counts=counts,
    color=Spectral10[:len(fruits)]
))

# Create figure
# Changed plot_height to height
p = figure(
    x_range=fruits,
    height=350,  # Use 'height' instead of 'plot_height'
    title="Fruit Count Visualization",
    toolbar_location=None
)

# Add bar chart
p.vbar(
    x='fruits',
    top='counts',
    width=0.9,
    source=source,
    line_color='white',
    fill_color='color'
)

# Customize appearance
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = 'Fruit Types'
p.yaxis.axis_label = 'Quantity'
p.title.text_font_size = '16pt'
p.title.align = 'center'

# Show the plot
show(p)

```

```python
# Que 4. Create a Bokeh histogram to visualize the distribution of the given data.

# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file

# Generate sample data (replace with your actual data)
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create figure
p = figure(
    title='Data Distribution Histogram',
    x_axis_label='Values',
    y_axis_label='Frequency',
    width=600,
    height=400
)

# Calculate histogram data
hist, edges = np.histogram(data, bins=30)

# Add histogram bars
p.quad(
    top=hist,
    bottom=0,
    left=edges[:-1],
    right=edges[1:],
    fill_color='skyblue',
    line_color='white'
)

# Output and show
output_file("histogram.html")
show(p)
```

```python
# Que 5. Create a Bokeh heatmap using the provided dataset.

#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.palettes import Viridis256

# Generate random heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create output file
output_file("heatmap.html")

# Create color mapper
color_mapper = LinearColorMapper(
    palette=Viridis256,
    low=data_heatmap.min(),
    high=data_heatmap.max()
)

# Create figure
p = figure(
    title='Random Heatmap',
    x_axis_label='X',
    y_axis_label='Y',
    width=500,
    height=500
)

# Add heatmap
p.image(
    [data_heatmap],
    x=0,
    y=0,
    dw=1,
    dh=1,
    color_mapper=color_mapper
)

# Add color bar
color_bar = ColorBar(
    color_mapper=color_mapper,
    location=(0, 0)
)
p.add_layout(color_bar, 'right')

# Show the plot
show(p)


```
